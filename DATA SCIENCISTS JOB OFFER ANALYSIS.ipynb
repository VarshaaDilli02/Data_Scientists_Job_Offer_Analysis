# Binomial Distribution

```python
#Probability of Data scientists whose company is of same state as job state
same_state=70/132

#Finding the probability for binomial distribution
def compute_binomial_probability(x, n, p):
    outcomes = math.factorial(n) / (math.factorial(x) * math.factorial(n - x))
    probability_of_each_outcome = p ** x * (1 - p) ** (n - x)
    return outcomes * probability_of_each_outcome
def plot_binomial_distribution_graph(n, p):
    probabilities = list(map(lambda x: compute_binomial_probability(x, n, p), range(0, n+1)))
    plt.bar(list(range(0, n+1)), probabilities)

#Plotting Binomial graph
plot_binomial_distribution_graph(132, same_state)
```

# Z - test

```python
# Statement of hypothesis
print("STATEMENT OF HYPOTHESIS")
print("Both Data Scientists and Data Engineers perform equally well w.r.t. average salary")
print("\n")

#Values found from the dataset
x1_bar=101.1301
x2_bar=100.6741
n1=131
n2=53
sig_1=38.82976
sig_2=38.82976

#Finding critical value for 95% confidence interval
def z_table(value):
   diff=1-(value/100)
   x=diff/2
   final=1-x
   p=norm.ppf(final)
   return p
print("Critical value of z:")
critical_z=z_table(95)
print(round(critical_z,2))

#Finding the test statistic
def z_test(x1_bar,x2_bar,n1,n2,sig_1,sig_2):
   num=(abs(x1_bar-x2_bar))
   dem=sqrt(((sig_1**2)/n1)+((sig_2**2)/n2))
   return num/dem
print("Test statistic of z:")
z_val=z_test(x1_bar,x2_bar,n1,n2,sig_1,sig_2)
print(round(z_val,2))

#Result of hypothesis
if z_val<critical_z: 
    print("We are accepting null hypothesis that mean average salary for 2 groups are equal")
else:
    print("We are rejecting null hypothesis that mean average salary for 2 groups are equal")
```

```python
#Comparing two distributions
def compare(arr1,arr2,alpha,sample_size):
    stat,p=ttest_ind(arr1,arr2)
    print('Statistics=%.3f,p=%.3f'%(stat,p))
    if p>alpha:
        print("Accepting null hypothesis")
        print("The datas are from Same distributions")
    else:
        print("Rejecting null hypothesis")
        print("The datas are from Different distributions")
```

```python
#Result of hypothesis
print("STATEMENT OF HYPOTHESIS")
print("Both the datas are from same distributions of average salary")
sample_size=15
test1=[55,61,61.5,77.5,61,61.5,77.5,87.5,138.5,94,87.5,138.5]
test2=[134,112.5,138.5,139,136.5,109,102.5,97,112.5,138.5,139,136.5,109,102.5]
s1=np.random.choice(test1,sample_size)
s2=np.random.choice(test2,sample_size)
print("Result:")
compare(s1,s2,0.05,sample_size)
```
